```python
import pandas as pd

data = {
    "resume_text": [
        "Python developer with machine learning, NLP and deep learning experience",
        "Java backend developer with Spring Boot, REST APIs and microservices",
        "Data scientist skilled in Python, pandas, scikit-learn and ML models",
        "Frontend developer with HTML, CSS, JavaScript and React experience",
        "AI engineer with Python, TensorFlow, deep learning and NLP projects"
    ],
    "job_role": [
        "ML Engineer",
        "Backend Developer",
        "Data Scientist",
        "Frontend Developer",
        "ML Engineer"
    ]
}

df = pd.DataFrame(data)

df.head()
```

```python
import re

def clean_text(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # remove numbers & punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra spaces
    return text

df["clean_resume"] = df["resume_text"].apply(clean_text)

df[["resume_text", "clean_resume"]].head()
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")


skills_list = [
    "python", "java", "machine learning", "deep learning",
    "nlp", "pandas", "scikit-learn", "tensorflow",
    "react", "spring boot", "sql", "html", "css", "javascript"
]

def extract_skills(text):
    text = text.lower()
    found_skills = []

    for skill in skills_list:
        if skill in text:
            found_skills.append(skill)

    return found_skills

df["skills"] = df["clean_resume"].apply(extract_skills)

df[["clean_resume", "skills"]].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df["clean_resume"])

print("Shape of TF-IDF matrix:", X.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity


job_descriptions = {
    "ML Engineer": "python machine learning deep learning nlp tensorflow",
    "Backend Developer": "java spring boot microservices api sql",
    "Data Scientist": "python pandas scikit-learn machine learning data analysis",
    "Frontend Developer": "html css javascript react frontend development"
}


df["job_description"] = df["job_role"].map(job_descriptions)


job_vectors = vectorizer.transform(df["job_description"])


df["match_score"] = cosine_similarity(X, job_vectors).diagonal()

df[["resume_text", "job_role", "match_score"]]
```

```python
ranked_df = df.sort_values(by="match_score", ascending=False)

ranked_df[["resume_text", "job_role", "match_score"]]
```
